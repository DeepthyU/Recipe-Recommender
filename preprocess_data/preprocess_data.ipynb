```python
val_dict={}
kcal_list=[]
fat_list=[]
saturates_list=[]
carbs_list=[]
sugars_list=[]
fibre_list=[]
protein_list=[]
salt_list=[]
for item in data1['nutrition_value']:
    if len(eval(item))>1:
        value=eval(item)
        kcal_list.append(value[0])
        fat_list.append(value[1].strip('g'))
        saturates_list.append(value[2].strip('g'))
        carbs_list.append(value[3].strip('g'))
        sugars_list.append(value[4].strip('g'))
        fibre_list.append(value[5].strip('g'))
        protein_list.append(value[6].strip('g'))
        salt_list.append(value[7].strip('g'))
    else:
        kcal_list.append(0)
        fat_list.append(0)
        saturates_list.append(0)
        carbs_list.append(0)
        sugars_list.append(0)
        fibre_list.append(0)
        protein_list.append(0)
        salt_list.append(0)
val_dict.update({'kcal':kcal_list, 'fat':fat_list, 'saturates':saturates_list, 'carbs':carbs_list, 'sugars':sugars_list,
                'fibre':fibre_list, 'protein':protein_list, 'salt':salt_list})
val_dict
```

```python
val=pd.DataFrame(val_dict)
val
```

```python
val.insert(loc=0,column='name',value=data1['name'])
val
```

```python
val.to_csv('clean_datasets/values.csv', index=False)
```

```python
name_list=[]
ingredient_list=[]
for i,item in enumerate(data1['ingredients']):
    for ing in eval(item):
        if ing not in ingredient_list:
            ingredient_list.append(ing)
            name_list.append(data1['name'][i])
for i,item in enumerate(data2['ingredients']):
    for ing in eval(item):
        if ing not in ingredient_list:
            ingredient_list.append(ing)
            name_list.append(data2['name'][i])
            
ing_dic={}
ing_dic.update({'name':name_list,'ingredients':ingredient_list})
pd.DataFrame(ing_dic)
ing_dic
```

```python
ings=pd.DataFrame(ing_dic)
ings
```

```python
ings.to_csv('clean_datasets/ingredients.csv', index=False)
```
